# Matplotlib Gallery Examples

- [The MatPlotLib gallery](http://matplotlib.org/gallery.html) exhibits hundreds of sample figures.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### Vertical Bar Chart

```python
X = np.arange(5,12)
X
```

```python
Y = X-7
Y
```

```python
plt.bar(X, Y)# Plot the bar chart
```

- How to determine the figure size and resolution?

```python
fig = plt.figure(figsize=(5, 4), dpi=80)
plt.bar(X, Y)
```

- How to display the values on bars? 

```python
fig = plt.figure(figsize=(5, 4), dpi=80)
plt.bar(X, Y)
for i in range(len(X)):
    if Y[i] >= 0:
        plt.text(X[i], Y[i] + 0.05, '{0:0.2f}'.format(Y[i]), horizontalalignment='center')
    else:
        plt.text(X[i], Y[i] - 0.3, '{0:0.2f}'.format(Y[i]), horizontalalignment='center')
plt.show()
```

- We need to overcome the overalp of values and y boundaries!

```python
fig = plt.figure(figsize=(5, 4), dpi=80)
plt.bar(X, Y)
for i in range(len(X)):
    if Y[i] >= 0:
        plt.text(X[i], Y[i] + 0.05, '{0:0.2f}'.format(Y[i]), horizontalalignment='center')
    else:
        plt.text(X[i], Y[i] - 0.3, '{0:0.2f}'.format(Y[i]), horizontalalignment='center')
plt.ylim(-3, 5)
plt.show()
```

### Horizontal Bar Chart

```python
cities = ['NYC', 'LA', 'Chicago']
indexes = np.arange(3)
wind = [10, 5, 13]  # Average wind speed in mph
error = [3, 1, 2]
fig = plt.figure(figsize=(5, 4), dpi=80)
plt.barh(indexes, wind, xerr=error, alpha=0.5)
plt.yticks(indexes, cities)
plt.xlabel('Wind Speed (mph)')
plt.title('Average Wind Speed at Three Cities')
plt.show()
```

### Boxplot

```python
data1 = np.random.normal(0, 1, 30)  # Normal distribution: =0, =1
data2 = np.random.normal(3, 2, 30)  # Normal distribution: =3, =2
fig = plt.figure(figsize=(5, 4), dpi=80)
plt.boxplot([data1, data2], showmeans=True)
plt.xticks([1, 2], ['data 1', 'data 2'])
plt.show()
```

### Pie Chart

```python
# The slices will be plotted counter-clockwise.
labels = ['Apples', 'Bananas', 'Pears', 'Oranges']
sizes = [25, 30, 50, 60]
colors = ['darkred', 'gold', 'greenyellow', 'darkorange']
explode = (0, 0.2, 0, 0)  # Only "explode" the 2nd slice
fig = plt.figure(figsize=(5, 4), dpi=80)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%0.1f%%', shadow=True, startangle=0)
plt.axis('equal')  # Set aspect ratio to be equal so the pie is a circle.
plt.show()
#autopct attribute is used to specify the distance of the text in the slice from the center of the circle.
```

### Histogram

```python
mu = 100  # Mean of distribution
sigma = 15  # Standard deviation (std)
x = mu + sigma * np.random.randn(10000)
num_bins = 30
fig = plt.figure(figsize=(5, 4), dpi=80)
n, bins, patches = plt.hist(x, num_bins, facecolor='green', edgecolor='k', alpha=0.5)
plt.xlabel('Value of x')
plt.ylabel('Count')
plt.title(r'$\mu=100, \sigma=15$')
plt.show()
```

### Wireframe (3D)

```python
from mpl_toolkits.mplot3d import axes3d
```

```python
fig = plt.figure(figsize=(5, 4), dpi=80)
ax = fig.add_subplot(111, projection='3d')
X, Y, Z = axes3d.get_test_data(0.1)
ax.plot_wireframe(X, Y, Z, rstride=5, cstride=5)  # rstride and cstride controls the row and col step sizes
ax.view_init(azim=170, elev=30)
plt.show()
```

### Surface (3D)

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(5, 4), dpi=80)
ax = fig.add_subplot(111, projection='3d')  # Use add_subplot to create a 3D subplot

# Create data points for X and Y within specified ranges and with specified step sizes
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)

# Create a meshgrid from X and Y
X, Y = np.meshgrid(X, Y)

# Define the Z values as the sine of the square root of the sum of squares of X and Y
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a surface plot with the specified properties
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0.2, antialiased=True)  # Use antialiased for smoother edges

# Set the z-axis limits
ax.set_zlim(-1.01, 1.01)

# Add a colorbar to the plot
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
# shrink: fraction by which to shrink the colorbar height
# aspect: ratio of height to width
```
